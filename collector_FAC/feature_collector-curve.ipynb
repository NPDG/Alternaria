```r
library(vegan)
```

```r
sort_1<-read.csv('sort1_bin_Alt_justPCR_EDITED_1000x.csv')
dim(sort_1)
```

```r
#12 columns of meta data
sort_1<-sort_1[,13:13146]
```

```r
sort_1_feat<-specaccum(sort_1, method="collector")
```

```r
plot(sort_1_feat,xlab='injection number',ylab='features',ci.type='poly',col='black',lwd=2,ci.lty=0,
     #ylim=c(0,12000),
     xlim=c(0,400),
    ci.col='gray',
    # main='Family Rarefaction in Alternaria'
    )
```
