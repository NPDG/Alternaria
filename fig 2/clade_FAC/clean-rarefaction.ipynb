```r
library(vegan)
```

## Clade U

```r
b_u<-read.csv('branch_u.csv')
dim(b_u)
```

```r
b_u<-b_u[,13:13146]
```

```r
rarespbu<-specaccum(b_u, method="rarefaction")
```

```r
specslope(rarespbu, at=221)
```

```r
plot(rarespbu,xlab='injection number',ylab='features',ci.type='poly',col='#4186f0',lwd=2,ci.lty=0,
     #ylim=c(0,12000),
     xlim=c(0,400),
    ci.col='lightgrey',
    # main='Rarefaction in Alternaria; Clade U'
    )
```
