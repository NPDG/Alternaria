# Central

```r
central<-read.csv('Central.csv')
dim(central)
```

```r
central<-central[,13:13146]
```

```r
rarecentral<-specaccum(central, method="rarefaction")
```

```r
specslope(rarecentral, at=29)
```

```r
plot(rarecentral,xlab='duplicate injection number',ylab='features',ci.type='poly',col='black',lwd=2,ci.lty=0,
     #ylim=c(0,12000),
     xlim=c(0,400),
    ci.col='gray',
    # main='Rarefaction in Alternaria; Central Region'
    )
```
