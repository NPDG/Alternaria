```r
library(vegan)
```

# branch_u-cluster_1

```r
c1_cu<-read.csv('c1_cu.csv')
dim(c1_cu)
```

```r
c1_cu<-c1_cu[,15:13148]
```

```r
rarec1_cu<-specaccum(c1_cu, method="rarefaction")
```

```r
specslope(rarec1_cu, at=203)
```

```r
plot(rarec1_cu,xlab='duplicate injection number',ylab='features',ci.type='poly',col='black',lwd=2,ci.lty=0,
     #ylim=c(0,12000),
     xlim=c(0,400),
    ci.col='gray',
    # main='Rarefaction in Alternaria; clade_u-cluster_1'
    )
```
