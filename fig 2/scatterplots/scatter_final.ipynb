```r
data<-read.csv('for_scatter.csv')
dim(data)
data
```

```r
x <- data$isolates
y<-data$n.1.slope.rarefaction
rarefaction_slope <- data$n.1.slope.rarefaction
plot(x,y)
```

```r
e_model<-lm(log(y)~x)
summary(e_model)
```

```r
x.values<-seq(0,400,1)
counts<-exp(predict(e_model,list(x=x.values)))
```

```r
plot(x, y, pch=19, col=c('black',"#4186f0","#1a237e",'#f8981b',"#097138","#9b27b0"),
    cex=2, xlab='injection number', ylab="rarefaction slope")
lines(x.values, counts, add=T)
```
