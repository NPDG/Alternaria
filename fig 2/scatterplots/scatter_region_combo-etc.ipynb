```r
data<-read.csv('for_scatter_combo-region.csv')
data
```

```r
regions <- data$count.of.regions
slope <- data$slope

plot(regions,slope, pch=19, 
     #col=c('black',"#4186f0","#1a237e",'#f8981b',"#097138","#9b27b0"),
    cex=2, xlab='regions')
```

```r
region<-read.csv('for_scatter_regions.csv')
region
```

```r
dup <- region$duplicates
slope <- region$slope

plot(dup,slope, pch=16, 
     #col=c('black',"#4186f0","#1a237e",'#f8981b',"#097138","#9b27b0"),
    cex=1.5, xlab='injection number')
```

```r
plot(dup,slope, pch=16, 
     #col=c('black',"#4186f0","#1a237e",'#f8981b',"#097138","#9b27b0"),
    cex=1.5, xlab='duplicate isolates')
```

```r
exponential.model <- lm(log(slope)~ dup)
summary(exponential.model)
```

```r
x.values<-seq(0,125,1)
counts<-exp(predict(exponential.model,list(x=x.values)))
```

```r
plot(dup,slope, pch=16, 
     #col=c('black',"#4186f0","#1a237e",'#f8981b',"#097138","#9b27b0"),
    cex=1.5, xlab='duplicate injection number')
lines(x.values, counts, add=T)
```
