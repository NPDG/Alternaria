```r
library(vegan)
```

```r
sort_1<-read.csv('sort_1_bin_summed_with_meta.csv')
dim(sort_1)
```

```r
#13 columns of meta data
sort_1<-sort_1[,14:1645]
```

```r
sort_1_fam<-specaccum(sort_1, method="collector")
```

```r
plot(sort_1_fam,xlab='injection number',ylab='chemical families',ci.type='poly',col='black',lwd=2,ci.lty=0,
     #ylim=c(0,12000),
     xlim=c(0,400),
    ci.col='gray',
    # main='Family Rarefaction in Alternaria'
    )
```
