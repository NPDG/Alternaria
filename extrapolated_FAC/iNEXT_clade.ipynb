## Extrapolated rarefaction curve of whole data set
#### endpoint should be set at 500

```r
# reading in transposed data 
all<-read.csv('t_bin_Alt.csv')
```

```r
# formatting data
mall <- as.matrix(apply(all[,-1],2,as.integer))
row.names(mall) <- all[,1]
head(mall)
```

```r
#create list
Alt = list(all = mall)
```

```r
# analyze and plot
out_all <- iNEXT(Alt, datatype="incidence_raw", endpoint=500)
ggiNEXT(out_all)
```

```r
#format appropriately 
l<-ggiNEXT(out_all, type=1)
l
l2 <- l + scale_colour_manual(values=c("black")) + theme(legend.position="bottom")+
scale_fill_manual(values=c("black")) + theme_bw(base_size = 12) +
 theme(legend.position="bottom", legend.title=element_blank())
l2
```

```r
# export values
out_all$iNextEst
```

## Extrapolated rarefaction curves of each clade
#### endpoint should be set at 500

```r
# read in data
by<-read.csv('use_tby.csv')
dim(by)
bx<-read.csv('use_tbx.csv')
dim(bx)
bw<-read.csv('use_tbw.csv')
dim(bw)
bv<-read.csv('use_tbv.csv')
dim(bv)
bu<-read.csv('use_tbu.csv')
dim(bu)
```

```r
# format data
mby<-as.matrix(apply(by[,-1],2,as.integer))
row.names(mby) <- by[,1]
mbx <- as.matrix(apply(bx[,-1],2,as.integer))
row.names(mbx) <- bx[,1]
mbw <- as.matrix(apply(bw[,-1],2,as.integer))
row.names(mbw) <- bw[,1]
mbv <- as.matrix(apply(bv[,-1],2,as.integer))
row.names(mbv) <- bv[,1]
mbu <- as.matrix(apply(bu[,-1],2,as.integer))
row.names(mbu) <- bu[,1]
```

```r
clade<-list(Y = mby, X = mbx, W = mbw, V = mbv, U = mbu)
```

```r
out_clade <- iNEXT(clade, datatype="incidence_raw", endpoint=500)
ggiNEXT(out_clade)
```

```r
#format figure
p<-ggiNEXT(out_clade, type=1)
p
p2 <- p + scale_colour_manual(values=c("#4186f0","#1a237e",'#f8981b',"#097138","#9b27b0")) + theme(legend.position="bottom")+
scale_fill_manual(values=c("#4186f0","#1a237e",'#f8981b',"#097138","#9b27b0")) + theme_bw(base_size = 12) +
 theme(legend.position="bottom", legend.title=element_blank())
p2
```

```r
# write values
out_clade$iNextEst
```
